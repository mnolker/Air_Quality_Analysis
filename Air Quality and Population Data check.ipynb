```python
import pandas as pd
```

```python
# Pull data from sites, aqi and census data to make sure there are no counties without data
sites_file = 'Data/ca_sites.csv'
sites_df = pd.read_csv(sites_file)

aqi_file = 'Data/aqi_full_data.csv'
aqi_df = pd.read_csv(aqi_file)
```

```python
# Census data file has to be cleaned up. The file starts as an Excel file with rows operating as header and footer.
census_county_file = 'Data/CA_Pop_Census_2010_2019_by_county.xlsx'
census_county = pd.read_excel(census_county_file, header=3, skipfooter=5)
census_county
```

```python
# Rename the first column
census_county.rename( columns={'Unnamed: 0':'County_state'}, inplace=True )
```

```python
# Parse the County_state column to keep only the county name
# First pass strips the "." at the beginning off
census_county[['Period', 'rest']] = census_county['County_state'].str.split(".", expand=True)

# Second pass creates County column with just column name
census_county[['County', 'rest2']] = census_county['rest'].str.split("County",expand=True)
```

```python
# Keep only the needed columns
columns = ['County', 'Census', 'Estimates Base', 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

census_df = census_county[columns]
census_df
```

```python
# Save to csv for future visualization.
census_df.to_csv("Data/census_info.csv")
```

```python
# Census data file has to be cleaned up. The file starts as an Excel file with rows operating as header and footer.
census_city_file = 'Data/CA_Pop_Census_2010_2019_by_city.xlsx'
census_city = pd.read_excel(census_city_file, header=3, skipfooter=5)

# Rename the first column
census_city.rename( columns={'Unnamed: 0':'City_state'}, inplace=True )
census_city
```

```python
# Create city column and clean the City_state column
census_city["City_state"] = census_city["City_state"].str.replace(" town", " city")
census_city
```

```python
census_city[['City', 'rest2']] = census_city['City_state'].str.split(" city",expand=True)
census_city
```

```python
# Save to csv for future visualization.
census_city.to_csv("Data/city_data.csv")
```
